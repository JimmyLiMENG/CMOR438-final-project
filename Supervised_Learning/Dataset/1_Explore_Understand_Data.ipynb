```python
df['tenure'].hist()
plt.title('Tenure Distribution')
plt.xlabel('Months')
plt.ylabel('Count')
```

> **Insight:**  
> - The overwhelming majority of customers fall within the first **0–7 months** of tenure, indicating a high risk of early churn.  
> - After month 7, the distribution levels off until a smaller secondary rise at the maximum tenure.

```python
df['Contract'].value_counts().plot(kind='bar')
plt.title('Contract Types')
plt.ylabel('Number of Customers')
```

> **Observation:**  
> - **Month-to-month** contracts dominate (≈3,800 customers).  
> - **Two-year** (≈1,700) and **One-year** (≈1,500) less common.

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')

# Select only numeric columns
numeric_df = df.select_dtypes(include=['number'])

corr = numeric_df.corr()
corr
```

**Quick take:**  
- `tenure` ↔ `TotalCharges` (r ≈ 0.83): longer tenure accumulates more charges.  
- `MonthlyCharges` ↔ `TotalCharges` (r ≈ 0.65): higher monthly spend → higher total.  
- `SeniorCitizen` shows low correlation with billing metrics.
